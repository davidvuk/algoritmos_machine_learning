# 0.Introduccion a Python



Python es un lenguaje de escritura de ***alto nivel, interpretado, imperativo, orientado a objetos, dinámico y fuerte*** que tiene las siguientes características:

* No hay declaración previa de variables, y los tipos de variables se       determinan dinámicamente.

* El control de bloques se realiza solo por identacion; No hay delimitadores de tipo BEGIN y END.

* Oferece tipos de datos de alto nível: strings, listas, tuplas, diccionarios, archivos, clases.

## 0.1 Esctructura de control

### 0.1.0 if y if-else

If ( ***Si se cumple esto pasa*** ) es una estructura de control tal que si se cumple la condicion realice la operaciones del interior.

If-Else( ***Si se cumple esto pasa al if , sino se cumple pasa a el else***) es una estructura de control tal que si se cumple la condicion realice la operaciones del interior al if , sino se cumple pasa al interior de el else.


```
# if 

if (1>2):
  print("4566")



if (1>2):
    print("4566")
else:
    print("hola clase")
```

### 0.1.1 For

For , es una ***estructura repetitiva*** que necesita un elemento iterador y una variable a cambiar.

```
import numpy as np


w=np.arange(5)
for i in w:
    print (i)

print (w)

ww="hola"
for i in ww:
    print (i)

print (type(ww))


```

### 0.1.2 While

While , es una estructura condicional , que necesita una variable booleana , ingresara hasta que la variable booleana cambie de valor.

```
year = 2001
while year <= 2012: 
    print ("Informes del Año " + str(year)) 
    year += 1
```

* Encontrar cuantas veces se repite la letra e en el texto " En esta clase se aprendera los conceptos basicos de programacion en python "

```
texto="En esta clase se aprendera los conceptos basicos de programacion en python"

cont=0
for i in texto:
    if (i=="e"):
        cont=cont+1

print(cont)
```
